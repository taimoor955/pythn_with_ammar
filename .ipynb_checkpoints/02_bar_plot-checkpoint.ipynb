```python
import seaborn as sns
import matplotlib.pyplot as plt
# data set
phool=sns.load_dataset("iris")
phool
# draw a line plot
sns.barplot(x="species", y="sepal_width", data=phool)
plt.title("Flower plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# load data set
kashti=sns.load_dataset("titanic")
kashti
# # draw a line plot
sns.barplot(x="sex", y="alone",hue="who", data=kashti, order=["female","male"], color="red",ci=None, palette="rocket")
plt.title("Titanic Jahaz")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median
# load data set
kashti=sns.load_dataset("titanic")
kashti
# # draw a line plot
sns.barplot(x="class", y="fare",hue="sex", data=kashti, color="red",estimator=median, palette="rocket")
plt.title("Titanic Jahaz")
plt.show()
```

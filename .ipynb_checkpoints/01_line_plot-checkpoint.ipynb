# Adding titles

```python
# import libarary
import seaborn as sns
import matplotlib.pyplot as plt
# data set
phool=sns.load_dataset("iris")
phool
# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Flower plot")
plt.show()
```

# Adding limits

```python
# import libarary
import seaborn as sns
import matplotlib.pyplot as plt
# data set
phool=sns.load_dataset("iris")
phool
# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.xlim(3)
plt.ylim(1)
plt.title("Flower plot")
plt.show()
```

# Size of Figure

```python
# import libarary
import seaborn as sns
import matplotlib.pyplot as plt
# change of figure
plt.figure(figsize=(4,3))
# data set
phool=sns.load_dataset("iris")
phool
# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Flower plot")
plt.show()
```
